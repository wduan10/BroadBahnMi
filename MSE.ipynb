```python
import os
import sys
import pandas as pd
import glob
from PIL import Image
import numpy as np
```

```python
label_path = '/groups/CS156b/data/student_labels/train2023.csv'
df = pd.read_csv(label_path)
df.head()
```

```python
matrix = np.array(df.iloc[:, -9:])
means = np.nanmean(matrix, axis=0)
means
```

```python
test_df = pd.read_csv('/groups/CS156b/data/student_labels/test_ids.csv')
test_df.head()
```

```python
rows_list = []
for i in range(len(test_df)):
    arr = [test_df['Id'][i]] + list(means)
    rows_list.append(arr)
df_output = pd.DataFrame(rows_list, columns=['Id'] + list(df.columns[-9:]))
df_output.head()
```

```python
output_dir = '/home/wduan'
full_path = os.path.join(output_dir, 'preds_MSE.csv')
df_output.to_csv(full_path, index=False)
```
