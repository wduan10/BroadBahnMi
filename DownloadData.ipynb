```python
import os
import sys
import pandas as pd
import glob
from PIL import Image
import shutil
```

## Train data

```python
label_path = '/groups/CS156b/data/student_labels/train2023.csv'
df = pd.read_csv(label_path)
df.head()
```

```python
image_paths = df['Path'][:-1]
image_paths_subset = image_paths[:25]
image_paths_subset
```

```python
prefix = '/groups/CS156b/data/train/'
dest_dir = '/groups/CS156b/2024/BroadBahnMi/data/train/images/'
for i, path in enumerate(image_paths_subset):
    pid = path.split('/')[1]
    fullpath = prefix + pid
    if (not os.path.isdir(dest_dir + pid)):
        print(fullpath, dest_dir)
        shutil.copytree(fullpath, dest_dir + pid)
```

```python
df = df[:100]
df
```

```python
df.to_csv('/groups/CS156b/2024/BroadBahnMi/data/train/labels/labels.csv', sep='\t')
```
